```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel ('job_data.xlsx', header=1)
```

```python
df.head()
```

```python
sns.barplot(x = "Industry", y = "Percent", data=df)
plt.title('Upper East Region: Employed population 15-35 years by industry', fontsize=7)
plt.tick_params(axis='both', which='major', labelsize=6)
plt.xticks(rotation=90)
```
