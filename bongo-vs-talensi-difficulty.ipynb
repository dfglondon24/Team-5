```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel ('difficulty_data.xlsx', header=1)
df = df.drop(0)
```

```python
df.head()
```

```python
sns.barplot(x = "Difficulty Status", y = "Bongo", data=df)
plt.title('Difficulty Status in Bongo', fontsize=7)
plt.tick_params(axis='both', which='major', labelsize=6)

```

```python
sns.barplot(x = "Difficulty Status", y = "Talensi", data=df)
plt.title('Difficulty Status in Talensi', fontsize=7)
plt.tick_params(axis='both', which='major', labelsize=6)

```
